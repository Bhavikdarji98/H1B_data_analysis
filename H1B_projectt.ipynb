```python
topdenied=df[df.case_status=='DENIED']['worksite'].value_counts()[:10]
```

```python
topdenied
```

```python
fig=plt.figure(figsize=(15,8),facecolor='mediumspringgreen')
x1=fig.add_axes([0,0,0.45,0.45])
x2=fig.add_axes([0.5,0,0.45,0.45])


x1.bar(x=topwork.index,height=topwork[:],color=col[::-1])
x1.set_xlabel('Worksites')
x1.set_ylabel('Number of Values')
x1.set_title('top 10 wprksites')
x1.set_xticklabels(labels=topwork.index,rotation=90)


x2.bar(x=topdenied.index,height=topdenied[:],color=col[::-3])
x2.set_title('top 10 denied worksites')
x1.set_xlabel('Worksites')
x2.set_ylabel('Number of Values')
x2.set_title('top 10 wprksites')
x2.set_xticklabels(labels=topdenied.index,rotation=90)

plt.show()
```

```python
job=['DATA SCIENTIST','DATA ANALYST','BUSINESS ANALYST','DATA ENGINEER']
total=[]
plt.rcParams['text.color']='white'
for var in job:
    d=df[df.job_title==var]['job_title'].count()
    total.append(d)
plt.figure(figsize=(10,5),facecolor='darkcyan')    
plt.bar(x=job[:],height=total[:],color=['plum','mediumvioletred','crimson','deepskyblue'])
plt.xlabel('Posts in job',color='white')
plt.ylabel('No. of jobs',color='white')
plt.title('Bar plot in various data jobs')
plt.show()
```
